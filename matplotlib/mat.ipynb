```python
import matplotlib.pyplot as plt
```

# Line plot

```python
x=[1,2,3,4,5]
y=[10,16,20,25,30]
plt.plot(x,y,color='red',marker='o',linestyle='-')
plt.title('lineplot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
time=['6am','9am','12pm','3pm','6pm']
temp=[18,22,28,30,24]
```

```python
plt.plot(time,temp,color='blue',marker='.',linestyle='-')
plt.title('lineplot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

# Bar plot

```python
names=['alice','bob','charlie','akku']
scores=[85,90,95,70]

plt.barh(names,scores,color='blue')
plt.title('Bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

# Histogram

```python
import numpy as np
data=np.random.randn(10000)
plt.hist(data,bins=30,color='purple')
plt.title('Histogram')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
import numpy as np
data=np.random.randint(1,100,size=500)
plt.hist(data,bins=30,color='purple')
plt.title('Histogram')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

# Scatter plot

```python
x=[5,7,8,7,2,17,2,9]
y=[99,86,87,88,100,86,103,87]
plt.scatter(x,y,color='red')
plt.title('scatter plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

# Pie chart

```python
labels=['python','java','c++','javascript']
sizes=[40,20,60,15]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=0)
plt.title('language popularity')
plt.axis('equal')
plt.show()
```
