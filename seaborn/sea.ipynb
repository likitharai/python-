```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('tips')
data.head()
```

#  line plot

```python
sns.lineplot(data=data,x='total_bill',y='tip')
plt.title('total bill')
plt.show()
```

# Bar plot

```python
sns.barplot(data=data,x='day',y='total_bill',estimator=sum)
plt.title('total bill collected per day')
plt.show()
```

```python
sns.barplot(data=data,x='day',y='total_bill',estimator='mean',palette='muted')
plt.title('total bill collected per day')
plt.show()
```

```python
sns.countplot(data=data,x='smoker',palette='muted')
plt.title('count')
```

# histogram

```python
sns.histplot(data['total_bill'],bins=20,kde=True,color='red')
plt.title('total bill')
plt.show()
```

```python
sns.histplot(data['tip'],bins=20,kde=True,color='red')
plt.title('total bill')
plt.show()
```

# Box plot

```python
sns.boxplot(data=data,x='day',y='total_bill',hue='sex')
plt.title('total bill collected per day')
plt.show()
```

```python
sns.boxplot(data=data,x='sex',y='total_bill')
plt.title('total bill collected per day')
plt.show()
```

```python
sns.violinplot(data=data,x='sex',y='total_bill')
plt.title('total bill collected per day')
plt.show()
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip',hue='sex',style='smoker')
plt.title('total bill collected per day')
plt.show()
```

# Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample: Load the CSV file you created (update path if needed)
data = pd.read_csv('tips.csv')

# Compute correlation matrix
correlation = data.corr()

# Plot heatmap
sns.heatmap(correlation, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```

# pairplot

```python
sns.pairplot(data,hue='sex')
```
