```python
import pandas as pd
url='https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv'
df=pd.read_csv(url)
```

```python
df.shape
```

```python
# handling missing values
df['Age']=df['Age'].fillna(df['Age'].median())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# remove duplicates
df=df.drop_duplicates()
```

```python
#filtering the data
first_class=df[df['Pclass']==1]
print(first_class.head(10))
```

```python
df.head(10)
```

```python
# generate visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
survived_by_pclass=df.groupby('Pclass')['Survived'].mean()
survived_by_pclass
```

```python
# bar chart for survived_by_pclass
survived_by_pclass.plot(kind='bar')
plt.xlabel('pclass')
plt.ylabel('survival rate')
plt.title('survived_by_pclass')
plt.show()
```

```python
#histogram age distribution
sns.histplot(df['Age'],bins=20,kde=True,color='red')
plt.title('age distribution')
plt.show()
```

```python
# scatter plot: age vs fare
plt.scatter(df['Age'],df['Fare'],alpha=0.5,color='green')
plt.title('age vs fare')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```
